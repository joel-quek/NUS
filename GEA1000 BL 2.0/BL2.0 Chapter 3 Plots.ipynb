```r
x <- seq(4, 8, length = 100)

# Create a linear relationship between x and y
y <- -x
e <- rnorm(100, 0, 0.2)
y <- y + e

# Set index and color for the outlier
outlier_index <- 50  # Index of the outlier point
outlier_color <- "red"  # Color for the outlier point

# Create vector for point colors
point_colors <- ifelse(1:length(x) == outlier_index, outlier_color, "black")

plot(x, y, pch = 16, col = point_colors, bg = point_colors)
```

```r
library(ggplot2)

# Generate data for clustered points
theta <- seq(0, 2*pi, length.out = 100)
radius <- 1
x_clustered <- radius * cos(theta)
y_clustered <- radius * sin(theta)

# Generate data for outlier point
x_outlier <- 4
y_outlier <- 4

# Create data frame
data <- data.frame(x = c(x_clustered, x_outlier),
                   y = c(y_clustered, y_outlier),
                   group = c(rep("Clustered", length(x_clustered)), "Outlier"))

# Create scatter plot
ggplot(data, aes(x, y, color = group)) +
  geom_point() +
  theme_bw()
```
