# Get List of Movie Websites from 2022

14008 movies in 2022

```python
page_num = np.arange(51,14050,50)
page_num
```

```python
url_links = ['https://www.imdb.com/search/title/?year=2022&title_type=feature&']

for page_ in page_num:
    url_ = "https://www.imdb.com/search/title/?title_type=feature&year="+str(2022)+"-01-01,"+str(2022)+"-12-31&start="+str(page_)+"&ref_=adv_nxt"
    url_links.append(url_)

url_links
```

```python
wd.get('https://www.imdb.com/search/title/?year=2022&title_type=feature&')
```

```python
wd.find_element("xpath","//*[@id='main']/div/div[3]/div/div[1]/div[3]/h3/a").get_attribute('href')
```

```python
wd.find_element("xpath","//*[@id='main']/div/div[3]/div/div[2]/div[3]/h3/a").get_attribute('href')
```

```python
movie_links = []

for url_ in url_links:
    wd.get(url_)

    try:
        for i in np.arange(1,51,1):
            movie_link = wd.find_element("xpath","//*[@id='main']/div/div[3]/div/div["+str(i)+"]/div[3]/h3/a").get_attribute('href')
            movie_links.append(movie_link)
    except:
        pass

movie_links
```

```python
movie_links
```

```python
with open('movielinks2.txt', 'w') as f:
    for line in movie_links:
        f.write(line)
        # f.write(",")
        f.write('\n')
```

Open Movie Links Text File into a List

```python
file = open("movielinks.txt", "r")

movie_list = []
for line in file:
  stripped_line = line.rstrip(line[-2:])
  movie_list.append(stripped_line)

movie_list
```
