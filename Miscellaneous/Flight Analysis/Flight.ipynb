# Opening BZ2 Files

Source: https://gist.github.com/yassineAlouini/86b19268b61c06df4e5c38dfb3e2e21c

Source 2: https://s3fs.readthedocs.io/en/latest/

```python
import s3fs
import pandas as pd

def get_csv_from_s3(folder_path):
    dfs = []
    s3 = s3fs.S3FileSystem()
    for fp in s3.ls(folder_path):
        if '.csv' in fp:
            with s3.open(fp) as s3f:
                with bz2file.open(s3f) as f:
                    tmp_df = pd.read_csv(f, encoding='utf-8', sep=';')
            dfs.append(tmp_df)
    if len(dfs) > 0:
        df = pd.concat(dfs)
        return df
```
